## **Load the dataset into a Pandas DataFrame and display the first 10 rows**

```python
df = pd.read_csv('/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv')
df.head()
```

```python
rows, columns = df.shape
print(f'{rows} Rows and {columns} Columns')
```

## **Select the Name, Age, and Salary columns using df[].**

```python
df[['Name', 'Age', 'Salary']].head()
```

## **Select the first 5 rows of the DataFrame using the iloc method**

```python
df.iloc[:5, :]
```

## **Select all rows where the Department is "Sales" using the loc method.**

```python
df.loc[df['Department'] == 'Sales']
```

## **Select rows where the Salary is greater than 8000 and Status is "Active".**

```python
df.loc[(df['Salary'] > 8000) & (df['Status'] == 'Active')]
```

## **Extract the rows where the employee has been with the company for more than 5 years and has a Performance Score of at least 4.**

```python
import datetime

df['Joining Date'] = pd.to_datetime(df['Joining Date'])
today = datetime.datetime.now()
df['Years In Service'] = (today - df['Joining Date']).dt.days // 365

df.loc[(df['Years In Service'] > 5) & (df['Performance Score'] >= 4)]
```

## **Select every alternate row from the DataFrame using slicing with iloc**

```python
df.iloc[::2,:]
```

## **Retrieve rows where the employee belongs to "Sales" and has a Salary in the top 10% of all salaries.**

```python
top10percnt = df['Salary'].quantile(0.9)
df.loc[(df['Department'] == 'Sales') & (df['Salary'] >= top10percnt)].head()
```

## **Replace all "Inactive" values in the Status column with "Retired".**

```python
df['Status'].replace({'Inactive':'Retired'}, inplace=True)
df['Status'].head()
```

## **Add a new column Bonus where values are 10% of the Salary column.**

```python
df['Bonus'] = df['Salary'] + (df['Salary'] * 0.10)
df['Bonus'].head()
```

## **Update the Performance Score of employees in the "IT" department to 5 if it is NaN.**

```python
it_df = df[df['Department'] == 'IT']
it_df['Performance Score'].replace({np.nan:5.0}, inplace=True)
it_df['Performance Score'].isna().sum()
it_df
```

## **Modify the Experience of employees with less than 2 years to 2.**

```python
df['Experience'] = df['Experience'].apply(lambda x: 2 if x<2 else x)
df.head()
```

## **Normalize the Salary column so that all values are between 0 and 1.**

```python
df['Experience'] = (df['Experience'] - df['Experience'].min()) / (df['Experience'].max() - df['Experience'].min())
df['Experience']
```

## **Replace all Location values with "Remote" for employees with a Performance Score below 3.**

```python
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
df.loc[:,['Performance Score', 'Location']].head(10)
```

## **Create a new column Experience Level with values "Junior", "Mid", or "Senior" based on the Experience.**

```python
df['Experience Level'] = df['Experience'].apply(lambda x: "Junior" if x <0.5 else "Mid" if (x >=0.5 and x<0.7) else "Senior")
df.loc[:,['Experience Level', 'Experience']].head(10)
```

## **Add a new row with specific values using the append() method.**

```python
new_row = pd.DataFrame({'ID':1001,'Name':'Ali Raza', 'Age':26, 'Gender':'Male', 'Department':'IT',
                       'Salary':'10000','Joining Date':'2023-10-14', 'Performance Score':4.0, 'Experience':4,
                       'Status':'Active','Location':'Moscow', 'Session':'Morning'}, index=[1001])
df = df._append(new_row)
df.tail()
```

## **Delete the column Session from the DataFrame.**

```python
df = df.drop('Session', axis=1)
df.head()
```

## **Delete rows where the Age is greater than 60.**

```python
df = df[df['Age'] <= 60]
df.head()
```

## **Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age.**

```python
df['Age Group'] = df['Age'].apply(lambda x: "Young" if x <=25 else "Middle-Aged" if (x >25 and x<=45) else "Senior")
df.loc[:, ['Age', 'Age Group']].head(10)
```

## **Identify all columns with missing values using the isna() method.**

```python
df.isna().sum()
```

## **Fill the missing values in Performance Score with the average score of that column.**

```python
mean_score = round(df['Performance Score'].mean(), 0)
df['Performance Score'].fillna(mean_score, inplace=True)
df.loc[:, ['Performance Score']]
```

## **Drop all rows with missing values using the dropna() method.**

```python
df = df.dropna()
df
```

## **Identify rows where the employee is in the "IT" department and their Age is above the average Age of "IT" employees**

```python
average_age_it = df.loc[df['Department'] == 'IT', 'Age'].mean()
df_it = df[df['Department'] == 'IT']
df_it[df_it['Age'] >= average_age_it]
```

## **Identify rows with duplicate Name and Joining Date combinations, and remove duplicates.**

```python
df = df.drop_duplicates(subset=['Name', 'Joining Date'])
df
```

## **Remove rows where the Experience value is inconsistent with the Age (e.g., Experience > Age - 18).**

```python
df.drop(index = df[df['Experience'] > df['Age']-18].index, axis = 0, inplace = True)
df
```

## **Sort the DataFrame by Salary in descending order.**

```python
sorted_df = df.sort_values(by='Salary', ascending=[False])
sorted_df
```

## **Sort the DataFrame by Department and then by Joining Date in ascending order.**

```python
sorted_df = df.sort_values(by=['Department', 'Joining Date'])
sorted_df
```

## **Rank all employees by Salary within their Department and add the rank as a new column.**

```python
sorted_df = df.sort_values(by=['Department', 'Salary'], ascending=[True, False])
sorted_df['Rank'] = sorted_df.groupby('Department').cumcount()
sorted_df
```

## **Sort the dataset by Experience in descending order, but break ties using Performance Score in ascending order.**

```python
sorted_df = df.sort_values(by=['Experience', 'Performance Score'])
sorted_df
```

## **Create a new DataFrame containing only employees in "HR" with a Salary above 5000.**

```python
hr_df = df[df['Department'] == 'HR']
final_df = hr_df[hr_df['Salary'] > 5000]
final_df.head()
```

## **Filter out all rows where the Location is "Chicago".**

```python
chicago_df = df[df['Location'] == "Chicago"]
chicago_df.head()
```

## **Delete all rows where the Performance Score is below 2.**

```python
df.drop(index = df[df['Performance Score'] < 2].index, axis = 0, inplace = True)
df.head()
```

## **Find the average Salary for each Department.**

```python
salary_df = df.groupby('Department')['Salary'].mean().reset_index()
salary_df
```

## **Calculate the total Salary of all employees grouped by Status.**

```python
total_salary_df = df.groupby('Status')['Salary'].sum().reset_index()
total_salary_df
```

## **Find the maximum Performance Score for employees in each Location.**

```python
max_score_df = df.groupby('Location')['Performance Score'].max().reset_index()
max_score_df
```

## **Calculate the total Salary paid to employees who joined in each year.**

```python
df['Joining Date']  = pd.to_datetime(df['Joining Date'])
df['Joining Year'] = df['Joining Date'].dt.year
year_wise_salary_df = df.groupby('Joining Year')['Salary'].sum().reset_index()
year_wise_salary_df
```

## **Find the department with the highest average Performance Score.**

```python
average_performance_score = df.groupby('Department')['Performance Score'].mean()

top_department = average_performance_score.idxmax()
top_department
```

## **Group employees by Status and calculate the sum, mean, and standard deviation of their Salary.**

```python
aggregate_sal = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
aggregate_sal
```

## **Use the apply() method to create a new column Monthly Salary by dividing the Salary by 12.**

```python
df['Monthly Salary'] = df['Salary'].apply(lambda x: x//12)
df
```

## **Use the applymap() method to convert all string values in the DataFrame to uppercase.**

```python
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
df
```

## **Create a second dataset containing Department and Budget. Perform an inner join with the original dataset.**

```python
dept_df = pd.DataFrame({'Department':['IT', 'HR', 'SALES'], 'Budget':[100000000, 5000000, 9000000]}, index=[0,1,2])
merged_df = pd.merge(df, dept_df, on='Department', how='inner')
merged_df
```

## **Merge the dataset with another dataset that includes additional information about employee ID and their recent Projects.**

```python
emp_df = pd.DataFrame({'ID':[1,2,3,4], 'Project':['Texas Floor Mill', 'Arizona National Museum', 'Philadephia Central Mall', 'Seattle International Airport']})
new_df = pd.merge(df, emp_df, on='ID', how='inner')
new_df.loc[:,['ID', 'Name', 'Project']]
```

## **Perform a left join with a dataset containing Department and Team Lead details, adding Team Lead to the DataFrame.**

```python
team_lead_df = pd.DataFrame({'Department': ['IT', 'HR', 'SALES'], 'Team Lead':['AMBER HARRIS', 'CORY ESCOBAR', 'TODD JOHNSON']})
new_df = pd.merge(df, team_lead_df, how='left', on='Department')
new_df.tail(6)
```

## **Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. Calculate the average Salary per project.**

```python
sub_df = df.loc[:5,:]
project_count_df = pd.DataFrame({'ID':[1,2,3,4,5], 'Project Count':[2,4,1,5,2]})
merged_df = pd.merge(sub_df, project_count_df, how='inner', on='ID')
merged_df['Avg Salary / Project'] = merged_df['Salary'] / merged_df['Project Count']
merged_df.loc[:, ['ID', 'Name', 'Salary', 'Project Count', 'Avg Salary / Project']]
```
